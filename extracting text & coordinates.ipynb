```python
import os
import re
```

```python
pdfpath='C:/Users/gangeshvar.s/Downloads/AI_11_ISC_2 1.pdf'
def extract_pages(pdfpath,lib):
    all_txt=[]
    if os.path.exists(pdfpath):         #to check whether there is the particular path in the system
        if lib == 'fitz' :
            try:
                doc=fitz.open(pdfpath)
                print(f'length of the document is {len(doc)}')
                for page in doc:
                    txt=page.get_text()
                    all_txt.append(re.sub(r"[^A-Za-z0-9 ]","",txt))    #^ dont match these items  r considers /s /n as the paths itself  
                                                                           # ' ' space will include space but will remove tags lik /n
                doc.close()
                return {"texts":all_txt}
            
            except Exception as e:
                print(e)

        elif lib == 'pdfplumber':
            try:
                with pdfplumber.open(pdfpath) as pdf:
                    print(f'length of the document is {len(pdf.pages)}')
                    for page in pdf.pages:
                        txt=page.extract_text()
                        all_txt.append(re.sub(r"[^A-Za-z0-9 ]","",txt))
                    pdf.close()
                    return {"texts":all_txt}
            
            except Exception as e:
                print(e)

        else:
            print('define library')
                
    else:
        print('file does not exist in the system')

datas = extract_pages(pdfpath,'pdfplumber')
```

```python
import fitz  # PyMuPDF

# Open PDF
doc = fitz.open(pdfpath)

# Loop through pages
for page_num in range(len(doc)):
    page = doc[page_num]
    
    # Extract text with positions
    words = page.get_text("words")  # returns list of tuples: (x0, y0, x1, y1, "word", block_no, line_no, word_no)
    
    for w in words:
        x0, y0, x1, y1, word = w[:5]
        print(f"Word: '{word}' at coordinates: ({x0}, {y0}, {x1}, {y1})")
```

```python
#extracting coordiates using pdfplumber
with pdfplumber.open(pdfpath) as pdf:
    for page in pdf.pages:
        txt=page.extract_words()         #extract_words extracts all the coordinates and props whereas and produces a dictionary 
        for w in txt:                    #extract_text extracts only the text and gives a list
            text=w['text']
            x0=w['x0']
            y0=w['top']
            x1=w['x1']
            y1=w['bottom']
            print(f"Word: '{text}' at coordinates: ({x0}, {y0}, {x1}, {y1})")
```
