```python
from typing import List, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values : List[int]
    name : str
    result : str
    operation : str
```

```python
import math

def addormul(state: AgentState) -> AgentState :
    if state["operation"] == "+" :
        state["result"] = f"Hi {state['name']}, your answer is: {sum(state['values'])}"
    elif state["operation"] == "*" :
        state["result"] = f"Hi {state['name']}, your answer is: {math.prod(state['values'])}"

    return state
```

```python
graph=StateGraph(AgentState)

graph.add_node("addormul",addormul)
graph.set_entry_point("addormul")
graph.set_finish_point("addormul")

app=graph.compile()
```

```python
result=app.invoke({"values":[1,2,3,4,5],"name":"ramesh","operation":"*"})
result
```
